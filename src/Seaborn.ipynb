# Seaborn Tutorial

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

Problem Statement:  We need to generate a model to predict tip based on other parameters or features.  Tip is a dependent feature where as others are independent features.  

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
```

# JoinPlot

It allows to study the relationship between 2 numeric variables.  This is also called bi-variant analysis.  

```python
sns.jointplot(data=df, x='total_bill', y='tip', kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='reg')
```

# Pair Plot

A pair plot is also known as a scatterplot, in which one variable is the same data set set row is matched with another variables value.  

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='time')
```

```python
sns.histplot(data=df, x='total_bill', kde=True)
```

```python
sns.countplot(x='time', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
```

```python
sns.boxplot(x='sex', y='total_bill', data=df)
```
