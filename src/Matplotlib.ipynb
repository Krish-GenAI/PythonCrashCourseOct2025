# Matplot lib Tutorial. 

Matplotlib is a plotting library for the python programming language and it is built for numerical mathematical extension on NumPy.  


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
plt.scatter(x,y,label='scatter plot',c='g')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('My first scatter plot')
plt.savefig('scatter.png')
```

```python
y = x*x
```

```python
plt.plot(x,y,'r',linestyle='dashed',linewidth=2,markersize=12,label='line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('My first line plot')
```

```python
# subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b*')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
# y = mx+c

x = np.arange(1,11)
y = (2*x) + 5
plt.title('Matplot lib demo')
plt.xlabel('X axis')
plt.ylabel('Y axis')    
plt.plot(x,y)
```

```python
# Sine and Cos ine Waves

x = np.arange(0,4*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x,y_sin,label='Sine Wave',color='r')
plt.plot(x,y_cos,label='Cosine Wave',color='b')
```

```python
plt.subplot(2,1,1)
plt.plot(x,y_sin,label='Sine Wave',color='r')
plt.subplot(2,1,2)
plt.plot(x,y_cos,label='Cosine Wave',color='b')
```

```python
# Bar Plot

x = [2,8,10]
y = [11,16,5]

x2 = [3,9,6]
y2 = [6,15,7]

plt.bar(x,y,label='Bars1',color='g')
plt.bar(x2,y2,label='Bars2',color='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Bar Plot')
```

# Histograms 

A histogram in python is a graphical representation of the distribution of a numerical variable in the assigned input array. 

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) # calculation for histogram. 

plt.hist(a, bins=10, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
```

# Pie Chart

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]    
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.0, 0, 0.3, 0)  # explode

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart')
```

# Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data,vert=True,patch_artist=True)
```
