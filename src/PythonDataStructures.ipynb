## Lists

List is mutable, changeable, it will have a specific sequence.  

```python
type([])
```

```python
lst_example = [1, 2, 3, 4, 5]
print(type(lst_example))  # <class 'list'>
```

```python
lst_multi_type = [1, "Hello", 3.14, True]
print(type(lst_multi_type))  # <class 'list'>
print(lst_multi_type)  # [1, 'Hello', 3.14, True
```

```python
len(lst_multi_type)  # 4
```

```python
lst_multi_type.append("New Element")
print(lst_multi_type)  # [1, 'Hello', 3.14, True
```

```python
lst_multi_type[0]
```

```python
lst_multi_type[4]
```

```python
lst_multi_type[2:4]
```

## Insert

```python
lst_multi_type.insert(2, "Inserted Element")
print(lst_multi_type)  # [1, 'Hello', 'Inserted Element', 3
```

```python
lst_multi_type.append("New Element2")
print(lst_multi_type)  # [1, 'Hello', 'Inserted Element', 3.14, True, 'New Element2']
```

```python
new_lst = ['Hello', 'World']
```

```python
lst_multi_type.append(new_lst)
```

```python
print(lst_multi_type)  # [1, 'Hello', 'Inserted Element', 3.14, True, 'New Element2', ['Hello', 'World']]
```

## Count() - Calculates total occurrence of a given element in a list. 

```python
lst_count = [1, 2, 2, 3, 4,4, 2, 5]
lst_count.count(2)  # 3
```

```python
len(lst_count)
```

```python
min(lst_count)  # 1
```

```python
max(lst_count)  # 5
```

## Sets - A set is an unordered collection that is iteratable, immutable and has no duplicate elements.  

```python
set_var = set()
print(set_var)
print(type(set_var))  # <class 'set'>
```

```python
set_var={1, 2, 3, 4, 5, 1, 2, 3}
```

```python
set_var
```

```python
set_var={"Avengers", "Batman", "Superman", "Spiderman", "Batman"}
set_var
```

```python
set_var.add("Hulk")
```

```python
set_var
```

```python
set1 = {"Apple", "Banana", "Cherry"}
set2 = {"Banana", "Dragonfruit", "Elderberry"}
```

```python
set2.intersection(set1)  # {'Banana'}
```

```python
set2.difference(set1)  # {'Dragonfruit', 'Elderberry'}
```

```python
set1.difference(set2)  # {'Apple', 'Cherry'}
```

```python
set1.difference_update(set2)
```

```python
set1
```

## Dictionaries

A dictionary is a collection which is unordered, changeable and indexed.  

```python
dic={}
print(type(dic))  # <class 'dict'>
```

```python
set_dict={"Name":"John", "Age":30, "City":"New York"}
print(set_dict)  # {'Name': 'John', 'Age': 30, '
```

```python
set_dict['Name']
```

```python
for x in set_dict:
    print(set_dict[x])
```

```python
for x in set_dict:
    print(x)
```

```python
for x in set_dict.values():
    print(x)
```

```python
for x in set_dict:
    print(x,set_dict[x])
```

```python
set_dict['Title']="Developer"
```

```python
for x in set_dict:
    print(x,set_dict[x])
```

## Nested Dictionary

```python
car1_model1 = {'Mecredes':2010}
car2_model2 = {'BMW':2015}
car3_model3 = {'Audi':2020}
```

```python
car_type = {
    'Car1':car1_model1,
    'Car2':car2_model2,
    'Car3':car3_model3}
```

```python
print(car_type)
```

```python
print(car_type['Car1'])
```

```python
print(car_type['Car1']['Mecredes'])
```
