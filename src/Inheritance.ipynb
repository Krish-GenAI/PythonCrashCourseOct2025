# Python OOPS - Inheritance

```python
class Car():
    def __init__(self,windows,doors,color,engineType):   #Constructor
        self.windows = windows
        self.doors = doors
        self.color = color
        self.engineType = engineType
    def drive(self): #Method
        print("The car is driving")
```

```python
car = Car(4,4,"Red","V8")
car.drive()
```

```python
# Extend the Car class to create a Audi class
class Audi(Car):
    def __init__(self, windows, doors, color, engineType, enableai):
        super().__init__(windows, doors, color, engineType)
        self.enableai = enableai

    def selfdriving(self):
        print("Audi supports self driving: ", self.enableai)
```

```python
audiq7 = Audi(4,4,"Black","V6",True)
audiq7.selfdriving() #Method from Audi class
audiq7.drive() #Inherited method from Car class
```
