Functions in Python.  A function is defined as a set of stetements that can reused in multiple scenarios to execute parameter based code and produces output.  

```python
## If Conditions

lst = [10, 20, 30, 40, 50, 5 , 7 , 17]

even_sum = 0
odd_sum = 0
for num in lst:
    if num % 2 == 0:
        even_sum += num
    else:
        odd_sum += num

print("Sum of even numbers:", even_sum)
print("Sum of odd numbers:", odd_sum)

```
